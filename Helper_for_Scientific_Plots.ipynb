In the following, we will plot three simple functions :

$$ 
 y=f(x) \text{ with} \begin{cases}
    y= \dfrac{x}{10}.\\
    y= \cos(x). \\
    y= \exp(-x).
  \end{cases} $$ 

First, we import the matploltib library and the numpy one

```python
import matplotlib.pyplot as plt
import numpy as np 
```

Then, we create our data

```python
Nb_points = 200
X_data  = np.linspace(0,30,Nb_points) 
Y_data  = [  X_data[i] / 10   for i in range(Nb_points) ]
Y_data2 = [ np.cos(X_data[i]) for i in range(Nb_points) ]
Y_data3 = [ np.exp(-X_data[i]) for i in range(Nb_points) ]
```

Now, we are going to plot these data with two different approaches : with and without  formatting.

## Using matplotlib with no formatting: BAD  !

```python
fig, ( ax1 ) = plt.subplots()
  
ax1.plot( X_data, Y_data, label='$y = x/10$')   
ax1.plot( X_data, Y_data2, ls='--',label='$y = \cos(x)$') 
ax1.plot( X_data, Y_data3, ls='-', label='$y = \exp(-x)$')

ax1.set_ylabel('Data for Y' )
ax1.set_xlabel('Data for X' )
ax1.legend() 

plt.tight_layout() 
plt.show()
```

<div class="alert alert-block alert-danger">
<b>This final product IS NOT GOOD to illustrate the results we obtained:</b> <br>
    - The curves  are <b>TOO THIN</b>.<br> 
    - The  text is <b>TOO SMALL</b> (on the axis, in the legend etc.). <br> 
    - It's <b>hard to follow the evolution of the curves</b> (amplitudes, variations etc.). <br> 
    <br> 
SO, to solve these problems we provide a slightly more elaborated canvas in the following part $\Downarrow$
</div>

****


## Using a custom template : BETTER !!! 

Here, we redefine the native parameters of the matlplolib library to solve the problem raied before and improve the quaility of the final plot. For more information, you can see the link 
https://matplotlib.org/stable/tutorials/introductory/customizing.html

```python
plt.rc('font',  family='serif')  # Type of font 
plt.rc('xtick', labelsize='24')  # Size of the xtick label 
plt.rc('ytick', labelsize='24')  # Size of the ytick label
plt.rc('lines', linewidth='3')   # Width of the curves  
plt.rc('axes', labelsize='26')   # Size of the Axes label (X and Y) 
plt.rc('legend', framealpha='1') # Transparency of the legend frame
plt.rc('legend', fontsize='20')  # Size of the legend
plt.rc('grid', linestyle='--')   # Grid formed by dashed lines 
plt.rcParams.update({ "text.usetex": True }) # Using LaTex style for text and equation
```

After redefining these parameters, the next plots produced will look like this

```python
fig, ( ax1 ) = plt.subplots( figsize=(8, 6) )
  
ax1.plot( X_data, Y_data, label='$y = x/10$')   
ax1.plot( X_data, Y_data2, ls='--', label='$y = \cos(x)$') 
ax1.plot( X_data, Y_data3, ls='-', label='$y = \exp(-x)$')
   
ax1.set_ylabel('Data for Y')
ax1.set_xlabel('Data for X') 
ax1.legend()  
ax1.grid()

plt.tight_layout() 
plt.savefig('MY_PLOT.pdf')

plt.show()
```

<div class="alert alert-block alert-success">
<b>This plot is in GOOD SHAPE to be used in scientific reports and  articles !</b>  
<br> 

Note that we exported this figure in ".PDF" by adding the line "plt.savefig('MY_PLOT.pdf')".

</div>

Sometimes, the last line invoking the use of LATEX style can be problematic. You can still use a simpler version of the template by turning off the last line of the template provided before :

```python
plt.rc('font',  family='Times')  # Type of font 
plt.rc('xtick', labelsize='24')  # Size of the xtick label 
plt.rc('ytick', labelsize='24')  # Size of the ytick label
plt.rc('lines', linewidth='3')   # Width of the curves  
plt.rc('axes', labelsize='26')   # Size of the Axes label (X and Y) 
plt.rc('legend', framealpha='1') # Transparency of the legend frame
plt.rc('legend', fontsize='20')  # Size of the legend 
plt.rc('grid', linestyle='--')   # Grid formed by dashed lines 
plt.rcParams.update({ "text.usetex": False }) # Using LaTex style for text and equation

fig, ( ax1 ) = plt.subplots( figsize=(8, 6) )
  
ax1.plot( X_data, Y_data, label='$y = x/10$')   
ax1.plot( X_data, Y_data2, ls='--', label='$y = \cos(x)$') 
ax1.plot( X_data, Y_data3, ls='-', label='$y = \exp(-x)$')
   
ax1.set_ylabel('Data for Y')
ax1.set_xlabel('Data for X') 
ax1.legend()  
ax1.grid()

plt.tight_layout() 
plt.savefig('MY_PLOT_2.pdf')

plt.show()
```
